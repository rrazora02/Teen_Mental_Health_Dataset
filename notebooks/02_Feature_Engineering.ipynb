```python
import pandas as pd
import numpy as np

from sklearn.model_selection import train_test_split
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline

from sklearn.preprocessing import (
    StandardScaler,
    OneHotEncoder
)

import joblib
```

```python
df = pd.read_csv("../data/cleaned_data.csv")

df.head()
```

```python
print(df.shape)

df.info()
```

```python
X = df.drop("depression_label", axis=1)

y = df["depression_label"]

print("Feature Shape :", X.shape)

print("Target Shape :", y.shape)
```

```python
print(y.value_counts())

print("\nPercentage\n")

print(y.value_counts(normalize=True) * 100)
```

```python
categorical_features = X.select_dtypes(include="object").columns.tolist()

numerical_features = X.select_dtypes(exclude="object").columns.tolist()

print("Categorical Features")

print(categorical_features)

print("\nNumerical Features")

print(numerical_features)
```

```python
numeric_transformer = Pipeline(
    steps=[
        ("scaler", StandardScaler())
    ]
)

categorical_transformer = Pipeline(
    steps=[
        ("encoder", OneHotEncoder(handle_unknown="ignore"))
    ]
)

preprocessor = ColumnTransformer(
    transformers=[
        ("num", numeric_transformer, numerical_features),
        ("cat", categorical_transformer, categorical_features)
    ]
)

preprocessor
```

```python
X_train, X_test, y_train, y_test = train_test_split(
    X,
    y,
    test_size=0.20,
    random_state=42,
    stratify=y
)

print("Training Set :", X_train.shape)

print("Testing Set :", X_test.shape)
```

```python
preprocessor.fit(X_train)
```

```python
X_train_processed = preprocessor.transform(X_train)

X_test_processed = preprocessor.transform(X_test)

print(X_train_processed.shape)

print(X_test_processed.shape)
```

```python
feature_names = preprocessor.get_feature_names_out()

print(feature_names)
```

```python
X_train_df = pd.DataFrame(
    X_train_processed,
    columns=feature_names
)

X_test_df = pd.DataFrame(
    X_test_processed,
    columns=feature_names
)

X_train_df.head()
```

```python
train_df = X_train_df.copy()

train_df["depression_label"] = y_train.reset_index(drop=True)

test_df = X_test_df.copy()

test_df["depression_label"] = y_test.reset_index(drop=True)

train_df.to_csv("../data/train_processed.csv", index=False)

test_df.to_csv("../data/test_processed.csv", index=False)

print("Processed datasets saved successfully!")
```

```python
joblib.dump(preprocessor, "../models/preprocessor.pkl")

print("Preprocessor saved successfully!")
```
