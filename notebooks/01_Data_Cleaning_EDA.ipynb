```python
plt.figure(figsize=(6,5))

sns.countplot(
    data=df,
    x="depression_label"
)

plt.title("Depression Distribution")
plt.show()
```

```python
plt.figure(figsize=(6,5))

sns.countplot(
    data=df,
    x="gender"
)

plt.title("Gender Distribution")
plt.show()
```

```python
plt.figure(figsize=(7,5))

sns.countplot(
    data=df,
    x="platform_usage"
)

plt.title("Most Used Platform")
plt.show()
```

```python
plt.figure(figsize=(8,5))

sns.histplot(
    df["age"],
    bins=10,
    kde=True
)

plt.title("Age Distribution")
plt.show()
```

```python
plt.figure(figsize=(8,5))

sns.histplot(
    df["daily_social_media_hours"],
    kde=True,
    bins=20
)

plt.title("Daily Social Media Usage")
plt.show()
```

```python
plt.figure(figsize=(8,5))

sns.histplot(
    df["sleep_hours"],
    bins=20,
    kde=True
)

plt.title("Sleep Hours Distribution")
plt.show()
```

```python
plt.figure(figsize=(8,5))

sns.boxplot(
    data=df,
    x="depression_label",
    y="stress_level"
)

plt.show()
```

```python
sns.pairplot(
    df[
        [
            "daily_social_media_hours",
            "sleep_hours",
            "stress_level",
            "anxiety_level",
            "addiction_level",
            "depression_label"
        ]
    ],
    hue="depression_label"
)

plt.show()
```
